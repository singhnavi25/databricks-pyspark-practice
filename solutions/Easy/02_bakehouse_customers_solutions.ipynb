# Solutions - Bakehouse Customers (Easy 02)

```python
from pyspark.sql import functions as F, types as T

df = spark.table("samples.bakehouse.sales_customers")
```

## Problem 1
Count customers per country, ordered descending.

```python
result_1 = df.groupBy("country").agg(
    F.count("*").alias("customer_count")
).orderBy(F.col("customer_count").desc())
result_1.show()
```

## Problem 2
Find all distinct continents.

```python
result_2 = df.select("continent").distinct()
result_2.show()
```

## Problem 3
Count customers per gender, replacing nulls with 'Unknown'.

```python
result_3 = df.fillna({"gender": "Unknown"}).groupBy("gender").agg(
    F.count("*").alias("count")
)
result_3.show()
```

## Problem 4
Find customers with null or empty state.

```python
result_4 = df.filter(
    F.col("state").isNull() | (F.trim(F.col("state")) == "")
).select("customerID", "first_name", "last_name", "country")
result_4.show()
```

## Problem 5
Find the top 3 cities by customer count.

```python
result_5 = df.groupBy("city", "country").agg(
    F.count("*").alias("customer_count")
).orderBy(F.col("customer_count").desc()).limit(3)
result_5.show()
```

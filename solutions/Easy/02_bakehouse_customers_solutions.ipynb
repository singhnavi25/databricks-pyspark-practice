# Solutions — Bakehouse Customers

**Dataset:** `samples.bakehouse.sales_customers`

**Topics:** groupBy, filter, distinct, null handling

> Reference solutions for practice notebooks.
> **Try the problem yourself first** — only open this when you've made a genuine attempt.
>
> Each solution includes a `# Why:` comment explaining the approach chosen.

```python
from pyspark.sql import functions as F, types as T
customers = spark.table("samples.bakehouse.sales_customers")
```

---
## Problem 1

```python
# Why: groupBy + count is the idiomatic way to get frequencies per group.
result_1 = (
    customers.groupBy("country")
             .count()
             .withColumnRenamed("count", "customer_count")
             .orderBy(F.col("customer_count").desc())
)
```

---
## Problem 2

```python
# Why: .select().distinct() is the cleanest way to list unique values.
result_2 = customers.select("continent").distinct().orderBy("continent")
```

---
## Problem 3

```python
# Why: coalesce returns the first non-null — ideal for a single-column null replacement.
result_3 = (
    customers.withColumn("gender", F.coalesce(F.col("gender"), F.lit("Unknown")))
             .groupBy("gender")
             .count()
)
```

---
## Problem 4

```python
# Why: use | (not Python 'or') to combine Spark column conditions.
result_4 = customers.filter(
    F.col("state").isNull() | (F.col("state") == "")
).select("customerID", "first_name", "last_name", "country", "state")
```

---
## Problem 5

```python
# Why: groupBy multiple columns simultaneously to count per city-country pair.
result_5 = (
    customers.groupBy("city", "country")
             .count()
             .withColumnRenamed("count", "customer_count")
             .orderBy(F.col("customer_count").desc())
             .limit(3)
)
```

# Solutions - Wanderbricks Bookings (Easy 08)


```python
from pyspark.sql import functions as F, types as T

df = spark.table("samples.wanderbricks.bookings")
```

## Problem 1
Count bookings per status. Columns: `status`, `booking_count`.


```python
result_1 = df.groupBy("status").agg(F.count("*").alias("booking_count"))
result_1.show()
```

## Problem 2
Revenue from confirmed bookings. Columns: `total_revenue`, `avg_booking_amount`.


```python
result_2 = df.filter(F.col("status") == "confirmed").agg(F.round(F.sum("total_amount"), 2).alias("total_revenue"), F.round(F.avg("total_amount"), 2).alias("avg_booking_amount"))
result_2.show()
```

## Problem 3
Bookings with more than 4 guests, sorted descending by guest count. All original columns.


```python
result_3 = df.filter(F.col("guests_count") > 4).orderBy(F.col("guests_count").desc())
result_3.show()
```

## Problem 4
Bookings per month. Columns: `year`, `month`, `booking_count`.


```python
result_4 = df.withColumn("year", F.year("check_in")).withColumn("month", F.month("check_in")).groupBy("year", "month").agg(F.count("*").alias("booking_count"))
result_4.show()
```

## Problem 5
Average nights per booking status. Columns: `status`, `avg_nights`.


```python
result_5 = df.withColumn("nights", F.datediff("check_out", "check_in")).groupBy("status").agg(F.round(F.avg("nights"), 1).alias("avg_nights"))
result_5.show()
```

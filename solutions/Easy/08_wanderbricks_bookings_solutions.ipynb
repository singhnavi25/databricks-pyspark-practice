```python
from pyspark.sql import functions as F, types as T
bookings = spark.table("samples.wanderbricks.bookings")
```

---
## Problem 1

```python
result_1 = (
    bookings.groupBy("status")
            .count()
            .withColumnRenamed("count", "booking_count")
)
```

---
## Problem 2

```python
result_2 = (
    bookings.filter(F.col("status") == "confirmed")
            .agg(
                F.round(F.sum("total_price"), 2).alias("total_revenue"),
                F.round(F.avg("total_price"), 2).alias("avg_booking_amount")
            )
)
```

---
## Problem 3

```python
result_3 = (
    bookings.filter(F.col("guests_count") > 4)
            .orderBy(F.col("guests_count").desc())
)
```

---
## Problem 4

```python
result_4 = (
    bookings.groupBy(
        F.year("check_in").alias("year"),
        F.month("check_in").alias("month")
    ).count()
     .withColumnRenamed("count", "booking_count")
     .orderBy("year", "month")
)
```

---
## Problem 5

```python
# Why: datediff(end, start) gives the number of days — for nights, check_out minus check_in.
result_5 = (
    bookings.groupBy("status")
            .agg(F.round(F.avg(F.datediff(F.col("check_out"), F.col("check_in"))), 1).alias("avg_nights"))
)
```

# Solutions - TPC-H Customer (Easy 05)

```python
from pyspark.sql import functions as F, types as T

df = spark.table("samples.tpch.customer")
```

## Problem 1
Count customers per market segment, ordered descending.

```python
result_1 = df.groupBy("c_mktsegment").agg(
    F.count("*").alias("customer_count")
).orderBy(F.col("customer_count").desc())
result_1.show()
```

## Problem 2
Find customers with positive account balance.

```python
result_2 = df.filter(F.col("c_acctbal") > 0).select(
    "c_custkey", "c_name", "c_acctbal"
)
result_2.show()
```

## Problem 3
Compute balance statistics by market segment.

```python
result_3 = df.groupBy("c_mktsegment").agg(
    F.round(F.avg("c_acctbal"), 2).alias("avg_balance"),
    F.min("c_acctbal").alias("min_balance"),
    F.max("c_acctbal").alias("max_balance")
)
result_3.show()
```

## Problem 4
Find customers with negative balance, sorted ascending.

```python
result_4 = df.filter(F.col("c_acctbal") < 0).select(
    "c_custkey", "c_name", "c_acctbal"
).orderBy("c_acctbal")
result_4.show()
```

## Problem 5
Find top 10 nation keys by customer count.

```python
result_5 = df.groupBy("c_nationkey").agg(
    F.count("*").alias("customer_count")
).orderBy(F.col("customer_count").desc()).limit(10)
result_5.show()
```

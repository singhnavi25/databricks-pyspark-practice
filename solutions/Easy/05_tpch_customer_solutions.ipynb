```python
from pyspark.sql import functions as F, types as T
customer = spark.table("samples.tpch.customer")
```

---
## Problem 1

```python
result_1 = (
    customer.groupBy("c_mktsegment")
            .count()
            .withColumnRenamed("count", "customer_count")
            .orderBy(F.col("customer_count").desc())
)
```

---
## Problem 2

```python
result_2 = (
    customer.filter(F.col("c_acctbal") > 0)
            .select("c_custkey", "c_name", "c_acctbal")
)
```

---
## Problem 3

```python
result_3 = (
    customer.groupBy("c_mktsegment")
            .agg(
                F.round(F.avg("c_acctbal"), 2).alias("avg_balance"),
                F.round(F.min("c_acctbal"), 2).alias("min_balance"),
                F.round(F.max("c_acctbal"), 2).alias("max_balance")
            )
)
```

---
## Problem 4

```python
result_4 = (
    customer.filter(F.col("c_acctbal") < 0)
            .select("c_custkey", "c_name", "c_acctbal")
            .orderBy("c_acctbal")
)
```

---
## Problem 5

```python
result_5 = (
    customer.groupBy("c_nationkey")
            .count()
            .withColumnRenamed("count", "customer_count")
            .orderBy(F.col("customer_count").desc())
            .limit(10)
)
```

# Solutions — Bakehouse Franchises

**Dataset:** `samples.bakehouse.sales_franchises`

**Topics:** groupBy, filter, distinct, count

> Each solution includes a `# Why:` comment explaining the approach chosen.

```python
from pyspark.sql import functions as F, types as T
franchises = spark.table("samples.bakehouse.sales_franchises")
```

---
## Problem 1

```python
result_1 = (
    franchises.groupBy("country")
              .count()
              .withColumnRenamed("count", "franchise_count")
              .orderBy(F.col("franchise_count").desc())
)
```

---
## Problem 2

```python
result_2 = franchises.select("size").distinct().orderBy("size")
```

---
## Problem 3

```python
result_3 = franchises.groupBy("size").count().orderBy(F.col("count").desc())
```

---
## Problem 4

```python
# Why: equality filter on a string column — case must match exactly.
result_4 = (
    franchises.filter(F.col("country") == "United Kingdom")
              .select("franchiseID", "name", "city", "size")
)
```

---
## Problem 5

```python
result_5 = (
    franchises.filter(F.col("district").isNull())
              .select("franchiseID", "name", "country")
)
```

# Solutions - Bakehouse Franchises (Easy 04)

```python
from pyspark.sql import functions as F, types as T

df = spark.table("samples.bakehouse.sales_franchises")
```

## Problem 1
Count franchises per country, ordered descending.

```python
result_1 = df.groupBy("country").agg(
    F.count("*").alias("franchise_count")
).orderBy(F.col("franchise_count").desc())
result_1.show()
```

## Problem 2
Find all distinct franchise sizes.

```python
result_2 = df.select("size").distinct()
result_2.show()
```

## Problem 3
Count franchises per size.

```python
result_3 = df.groupBy("size").agg(
    F.count("*").alias("count")
)
result_3.show()
```

## Problem 4
Filter franchises in the United Kingdom.

```python
result_4 = df.filter(F.col("country") == "United Kingdom").select(
    "franchiseID", "name", "city", "size"
)
result_4.show()
```

## Problem 5
Find franchises with null district.

```python
result_5 = df.filter(F.col("district").isNull()).select(
    "franchiseID", "name", "country"
)
result_5.show()
```

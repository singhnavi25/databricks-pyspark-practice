# Solutions - TPC-H Orders (Easy 06)


```python
from pyspark.sql import functions as F, types as T

df = spark.table("samples.tpch.orders")
```

## Problem 1
Count orders by status. Columns: `o_orderstatus`, `order_count`.


```python
result_1 = df.groupBy("o_orderstatus").agg(F.count("*").alias("order_count"))
result_1.show()
```

## Problem 2
Revenue per priority, descending. Columns: `o_orderpriority`, `total_revenue`.


```python
result_2 = df.groupBy("o_orderpriority").agg(F.sum("o_totalprice").alias("total_revenue")).orderBy(F.col("total_revenue").desc())
result_2.show()
```

## Problem 3
Count orders placed in 1995. Column: `order_count_1995`.


```python
result_3 = df.filter(F.year("o_orderdate") == 1995).agg(F.count("*").alias("order_count_1995"))
result_3.show()
```

## Problem 4
Monthly order volume. Columns: `year`, `month`, `order_count`, `avg_price`.


```python
result_4 = df.withColumn("year", F.year("o_orderdate")).withColumn("month", F.month("o_orderdate")).groupBy("year", "month").agg(F.count("*").alias("order_count"), F.round(F.avg("o_totalprice"), 2).alias("avg_price"))
result_4.show()
```

## Problem 5
Top 5 clerks by order count. Columns: `o_clerk`, `order_count`.


```python
result_5 = df.groupBy("o_clerk").agg(F.count("*").alias("order_count")).orderBy(F.col("order_count").desc()).limit(5)
result_5.show()
```

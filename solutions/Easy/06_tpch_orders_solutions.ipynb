**Topics:** filter, date functions, groupBy, aggregation

> Each solution includes a `# Why:` comment explaining the approach chosen.

```python
from pyspark.sql import functions as F, types as T
orders = spark.table("samples.tpch.orders")
```

---
## Problem 1

```python
result_1 = (
    orders.groupBy("o_orderstatus")
          .count()
          .withColumnRenamed("count", "order_count")
)
```

---
## Problem 2

```python
result_2 = (
    orders.groupBy("o_orderpriority")
          .agg(F.round(F.sum("o_totalprice"), 2).alias("total_revenue"))
          .orderBy(F.col("total_revenue").desc())
)
```

---
## Problem 3

```python
# Why: filter on the extracted year, then count — single-row result.
result_3 = (
    orders.filter(F.year("o_orderdate") == 1995)
          .agg(F.count("*").alias("order_count_1995"))
)
```

---
## Problem 4

```python
# Why: extract year and month simultaneously in one groupBy call.
result_4 = (
    orders.groupBy(
        F.year("o_orderdate").alias("year"),
        F.month("o_orderdate").alias("month")
    ).agg(
        F.count("*").alias("order_count"),
        F.round(F.avg("o_totalprice"), 2).alias("avg_price")
    ).orderBy("year", "month")
)
```

---
## Problem 5

```python
result_5 = (
    orders.groupBy("o_clerk")
          .count()
          .withColumnRenamed("count", "order_count")
          .orderBy(F.col("order_count").desc())
          .limit(5)
)
```

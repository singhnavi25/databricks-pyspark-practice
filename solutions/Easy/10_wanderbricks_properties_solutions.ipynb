# Solutions - Wanderbricks Properties (Easy 10)


```python
from pyspark.sql import functions as F, types as T

df = spark.table("samples.wanderbricks.properties")
```

## Problem 1
Count per property type, descending. Columns: `property_type`, `count`.


```python
result_1 = df.groupBy("property_type").agg(F.count("*").alias("count")).orderBy(F.col("count").desc())
result_1.show()
```

## Problem 2
Average price per property type, descending. Columns: `property_type`, `avg_price`.


```python
result_2 = df.groupBy("property_type").agg(F.round(F.avg("base_price"), 2).alias("avg_price")).orderBy(F.col("avg_price").desc())
result_2.show()
```

## Problem 3
Properties with 3 or more bedrooms. Columns: `property_id`, `title`, `property_type`, `bedrooms`, `base_price`.


```python
result_3 = df.filter(F.col("bedrooms") >= 3).select("property_id", "title", "property_type", "bedrooms", "base_price")
result_3.show()
```

## Problem 4
Top 10 hosts by property count. Columns: `host_id`, `property_count`.


```python
result_4 = df.groupBy("host_id").agg(F.count("*").alias("property_count")).orderBy(F.col("property_count").desc()).limit(10)
result_4.show()
```

## Problem 5
10 cheapest properties with at least 2 guests. Columns: `property_id`, `title`, `property_type`, `base_price`, `max_guests`.


```python
result_5 = df.filter(F.col("max_guests") >= 2).select("property_id", "title", "property_type", "base_price", "max_guests").orderBy("base_price").limit(10)
result_5.show()
```

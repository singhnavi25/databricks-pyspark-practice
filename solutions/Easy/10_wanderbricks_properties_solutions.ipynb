**Dataset:** `samples.wanderbricks.properties`

**Topics:** filter, aggregation, groupBy

```python
from pyspark.sql import functions as F, types as T
properties = spark.table("samples.wanderbricks.properties")
```

---
## Problem 1

```python
result_1 = (
    properties.groupBy("property_type")
              .count()
              .orderBy(F.col("count").desc())
)
```

---
## Problem 2

```python
result_2 = (
    properties.groupBy("property_type")
              .agg(F.round(F.avg("base_price"), 2).alias("avg_price"))
              .orderBy(F.col("avg_price").desc())
)
```

---
## Problem 3

```python
result_3 = (
    properties.filter(F.col("bedrooms") >= 3)
              .select("property_id", "title", "property_type", "bedrooms", "base_price")
)
```

---
## Problem 4

```python
result_4 = (
    properties.groupBy("host_id")
              .count()
              .withColumnRenamed("count", "property_count")
              .orderBy(F.col("property_count").desc())
              .limit(10)
)
```

---
## Problem 5

```python
result_5 = (
    properties.filter(F.col("max_guests") >= 2)
              .select("property_id", "title", "property_type", "base_price", "max_guests")
              .orderBy("base_price")
              .limit(10)
)
```

**Dataset:** `samples.wanderbricks.users`

**Topics:** filter, groupBy, boolean columns, date

```python
from pyspark.sql import functions as F, types as T
users = spark.table("samples.wanderbricks.users")
```

---
## Problem 1

```python
result_1 = (
    users.groupBy("country")
         .count()
         .withColumnRenamed("count", "user_count")
         .orderBy(F.col("user_count").desc())
         .limit(10)
)
```

---
## Problem 2

```python
# Why: boolean columns work directly in groupBy — produces True/False rows.
result_2 = (
    users.groupBy("is_business")
         .count()
         .withColumnRenamed("count", "user_count")
)
```

---
## Problem 3

```python
result_3 = (
    users.groupBy(F.year("created_at").alias("join_year"))
         .count()
         .withColumnRenamed("count", "user_count")
         .orderBy("join_year")
)
```

---
## Problem 4

```python
result_4 = (
    users.groupBy("user_type")
         .count()
         .orderBy(F.col("count").desc())
)
```

---
## Problem 5

```python
result_5 = (
    users.filter(F.col("company_name").isNotNull())
         .select("user_id", "name", "country", "company_name")
)
```

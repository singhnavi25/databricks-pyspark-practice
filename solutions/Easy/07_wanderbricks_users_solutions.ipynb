# Solutions - Wanderbricks Users (Easy 07)


```python
from pyspark.sql import functions as F, types as T

df = spark.table("samples.wanderbricks.users")
```

## Problem 1
Top 10 countries by user count. Columns: `country`, `user_count`.


```python
result_1 = df.groupBy("country").agg(F.count("*").alias("user_count")).orderBy(F.col("user_count").desc()).limit(10)
result_1.show()
```

## Problem 2
Business vs individual user count. Columns: `is_business`, `user_count`.


```python
result_2 = df.groupBy("is_business").agg(F.count("*").alias("user_count"))
result_2.show()
```

## Problem 3
Users per join year. Columns: `join_year`, `user_count`.


```python
result_3 = df.withColumn("join_year", F.year("created_at")).groupBy("join_year").agg(F.count("*").alias("user_count")).orderBy("join_year")
result_3.show()
```

## Problem 4
Count per user type. Columns: `user_type`, `count`.


```python
result_4 = df.groupBy("user_type").agg(F.count("*").alias("count"))
result_4.show()
```

## Problem 5
Users with a company name. Columns: `user_id`, `name`, `country`, `company_name`.


```python
result_5 = df.filter(F.col("company_name").isNotNull()).select("user_id", "name", "country", "company_name")
result_5.show()
```

# Solutions — TPC-H Nation & Region

**Dataset:** `samples.tpch.nation` + `samples.tpch.region`

**Topics:** join, groupBy, filter

> Reference solutions for practice notebooks.
> **Try the problem yourself first** — only open this when you've made a genuine attempt.
>
> Each solution includes a `# Why:` comment explaining the approach chosen.

```python
from pyspark.sql import functions as F, types as T
nations = spark.table("samples.tpch.nation")
regions = spark.table("samples.tpch.region")
```

---
## Problem 1

```python
# Why: join on the key expression (different column names) then select to avoid ambiguity.
result_1 = (
    nations.join(regions, nations.n_regionkey == regions.r_regionkey, "inner")
           .select("n_name", "r_name")
           .orderBy("r_name", "n_name")
)
```

---
## Problem 2

```python
result_2 = (
    nations.join(regions, nations.n_regionkey == regions.r_regionkey, "inner")
           .groupBy("r_name")
           .count()
           .withColumnRenamed("count", "nation_count")
           .orderBy("r_name")
)
```

---
## Problem 3

```python
result_3 = (
    nations.join(regions, nations.n_regionkey == regions.r_regionkey, "inner")
           .filter(F.col("r_name") == "AFRICA")
           .select("n_name")
           .orderBy("n_name")
)
```

---
## Problem 4

```python
result_4 = (
    nations.join(regions, nations.n_regionkey == regions.r_regionkey, "inner")
           .filter(F.col("r_name") == "EUROPE")
           .select("n_name")
           .orderBy("n_name")
)
```

---
## Problem 5

```python
# Why: collect_list gathers all group values into a single array per region.
result_5 = (
    nations.join(regions, nations.n_regionkey == regions.r_regionkey, "inner")
           .groupBy(F.col("r_name").alias("region"))
           .agg(F.collect_list("n_name").alias("nations"))
           .orderBy("region")
)
```

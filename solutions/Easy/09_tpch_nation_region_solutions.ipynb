# Solutions - TPC-H Nation & Region (Easy 09)


```python
from pyspark.sql import functions as F, types as T

nations = spark.table("samples.tpch.nation")
regions = spark.table("samples.tpch.region")
```

## Problem 1
Join nations with regions, sorted. Columns: `n_name`, `r_name`.


```python
result_1 = nations.join(regions, nations.n_regionkey == regions.r_regionkey).select("n_name", "r_name").orderBy("r_name", "n_name")
result_1.show()
```

## Problem 2
Nations per region. Columns: `r_name`, `nation_count`.


```python
result_2 = nations.join(regions, nations.n_regionkey == regions.r_regionkey).groupBy("r_name").agg(F.count("*").alias("nation_count"))
result_2.show()
```

## Problem 3
Nations in AFRICA. Column: `n_name`.


```python
result_3 = nations.join(regions, nations.n_regionkey == regions.r_regionkey).filter(F.col("r_name") == "AFRICA").select("n_name")
result_3.show()
```

## Problem 4
Nations in EUROPE. Column: `n_name`.


```python
result_4 = nations.join(regions, nations.n_regionkey == regions.r_regionkey).filter(F.col("r_name") == "EUROPE").select("n_name")
result_4.show()
```

## Problem 5
Collect nation names per region as array. Columns: `region`, `nations`.


```python
result_5 = nations.join(regions, nations.n_regionkey == regions.r_regionkey).groupBy(F.col("r_name").alias("region")).agg(F.collect_list("n_name").alias("nations"))
result_5.show()
```

**Dataset:** `samples.nyctaxi.trips`

**Topics:** filter, count, agg, sorting

> Each solution includes a `# Why:` comment explaining the approach chosen.

```python
from pyspark.sql import functions as F, types as T
df = spark.table("samples.nyctaxi.trips")
```

---
## Problem 1

```python
# Why: .count() returns a Python int, not a DataFrame.
# agg(F.count("*")) returns a single-row DataFrame with the count as a column.
result_1 = df.agg(F.count("*").alias("total_trips"))
```

---
## Problem 2

```python
# Why: filter() is a lazy transformation — no compute until an action is called.
# Returning all columns satisfies "return the full original row".
result_2 = df.filter(F.col("trip_distance") > 10)
```

---
## Problem 3

```python
# Why: pass multiple expressions to a single agg() to compute all stats in one pass.
result_3 = df.agg(
    F.round(F.avg("fare_amount"), 2).alias("avg_fare"),
    F.min("fare_amount").alias("min_fare"),
    F.max("fare_amount").alias("max_fare")
)
```

---
## Problem 4

```python
# Why: groupBy().count() then rename, sort, and limit — cleaner than agg(F.count("*")).
result_4 = (
    df.groupBy("pickup_zip")
      .count()
      .withColumnRenamed("count", "trip_count")
      .orderBy(F.col("trip_count").desc())
      .limit(5)
)
```

---
## Problem 5

```python
# Why: filter first to narrow the dataset, then groupBy — keeps the pipeline clear.
result_5 = (
    df.filter(F.col("fare_amount") > 20)
      .groupBy("dropoff_zip")
      .count()
      .withColumnRenamed("count", "trip_count")
)
```

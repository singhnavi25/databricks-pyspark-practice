# Solutions - NYC Taxi Trips Basics (Easy 01)

```python
from pyspark.sql import functions as F, types as T

df = spark.table("samples.nyctaxi.trips")
```

## Problem 1
Count total trips.

```python
result_1 = df.agg(F.count("*").alias("total_trips"))
result_1.show()
```

## Problem 2
Filter trips with distance greater than 10 miles.

```python
result_2 = df.filter(F.col("trip_distance") > 10).select(
    "tpep_pickup_datetime", "tpep_dropoff_datetime",
    "trip_distance", "fare_amount", "pickup_zip", "dropoff_zip"
)
result_2.show()
```

## Problem 3
Compute fare statistics: average, minimum, and maximum.

```python
result_3 = df.agg(
    F.round(F.avg("fare_amount"), 2).alias("avg_fare"),
    F.min("fare_amount").alias("min_fare"),
    F.max("fare_amount").alias("max_fare")
)
result_3.show()
```

## Problem 4
Find the top 5 pickup zip codes by trip count.

```python
result_4 = df.groupBy("pickup_zip").agg(
    F.count("*").alias("trip_count")
).orderBy(F.col("trip_count").desc()).limit(5)
result_4.show()
```

## Problem 5
Count trips per dropoff zip where fare exceeds $20.

```python
result_5 = df.filter(F.col("fare_amount") > 20).groupBy("dropoff_zip").agg(
    F.count("*").alias("trip_count")
)
result_5.show()
```

# Hints — Higher-Order Functions and JSON Parsing

**Level:** Medium

> These hints are here when you're stuck. Try solving each problem on your own first.
> Reveal one hint at a time — start with Problem 1 and work through in order.

## Problem 1

<details>
<summary>Hint 1</summary>

`F.transform(col, lambda x: ...)` applies a function to every element of an array column. Your lambda should double each element: `lambda x: x * 2`.

</details>

## Problem 2

<details>
<summary>Hint 1</summary>

The higher-order `F.filter()` takes an array column and a lambda predicate. Don't confuse it with `df.filter()` which filters rows. Use `F.filter(col, lambda x: x > 12)`.

</details>

## Problem 3

<details>
<summary>Hint 1</summary>

`F.aggregate(col, initial_value, merge_func)` folds an array into a single value. The initial value should be `F.lit(0)` and the merge function adds each element to the accumulator: `lambda acc, x: acc + x`.

</details>

## Problem 4

<details>
<summary>Hint 1</summary>

`F.exists(col, lambda x: ...)` returns True if any element satisfies the predicate. `F.forall(col, lambda x: ...)` returns True only if all elements satisfy it. To check for odd: `x % 2 != 0`. To check for even: `x % 2 == 0`.

</details>

## Problem 5

<details>
<summary>Hint 1</summary>

`F.zip_with(col1, col2, lambda x, y: ...)` pairs elements by index and applies the lambda. For element-wise multiplication use `lambda x, y: x * y`.

</details>

## Problem 6

<details>
<summary>Hint 1</summary>

Start with `.groupBy("franchiseID").agg(F.collect_set("product").alias("products"))`. Then chain `.withColumn` calls for `array_distinct`, `array_sort`, `array_size`, and `array_join`.

</details>

<details>
<summary>Hint 2</summary>

`F.array_join(col, ", ")` converts an array of strings into a single comma-separated string. Make sure you apply `array_sort` before `array_join` so the output is deterministic.

</details>

## Problem 7

<details>
<summary>Hint 1</summary>

Use `F.schema_of_json(F.lit('{"name":"Alice","age":30}'))` to infer the schema, then pass that schema to `F.from_json("json_str", schema)` to parse each row.

</details>

<details>
<summary>Hint 2</summary>

Once you have the `parsed` struct column, extract fields with `F.col("parsed.name").alias("name")` and `F.col("parsed.age").alias("age")`.

</details>

## Problem 8

<details>
<summary>Hint 1</summary>

`F.create_map(F.col("product"), F.col("totalPrice"))` builds a single-entry map per row. Then use `F.map_keys(col)` and `F.map_values(col)` to extract the keys and values as arrays.

</details>

## Problem 3


<details>
<summary>Hint 1 — add a computed total_length column before filtering on it</summary>

Build the length columns with `withColumn`, then filter using the new column:
```python
df.withColumn("first_name_length", F.length("first_name"))
  .withColumn("last_name_length", F.length("last_name"))
  .withColumn("total_length", F.col("first_name_length") + F.col("last_name_length"))
  .filter(F.col("total_length") > 20)
```

## Problem 1


<details>
<summary>Hint 1 — which function counts rows and returns a DataFrame?</summary>

`df.count()` is an **action** that returns a Python int, not a DataFrame.

To get a single-row DataFrame with the count as a column, use:
```python
df.agg(F.count("*").alias("total_trips"))
```

</details>


## Problem 3


<details>
<summary>Hint 1 — aggregating multiple stats in one call</summary>

You can pass multiple aggregate expressions to a single `.agg()` call:
```python
df.agg(
    F.avg("fare_amount").alias("avg_fare"),
    F.min("fare_amount").alias("min_fare"),
    F.max("fare_amount").alias("max_fare")
)
```

</details>


## Problem 4


<details>
<summary>Hint 1 — getting exactly 5 rows in a specific order</summary>

Chain `.orderBy()` and `.limit()` together:
```python
df.groupBy("pickup_zip") \
  .count() \
  .orderBy(F.col("count").desc()) \
  .limit(5)
```
Rename the `count` column to `trip_count` using `.withColumnRenamed("count", "trip_count")`.

</details>


<details>
<summary>Hint 2 — renaming the default count column</summary>

`.count()` produces a column named `count`. You can either rename it after:
```python
.withColumnRenamed("count", "trip_count")
```
Or use `.agg(F.count("*").alias("trip_count"))` instead of `.count()`.

</details>

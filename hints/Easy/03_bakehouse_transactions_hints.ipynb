## Problem 6


<details>
<summary>Hint 1 — finding the min and max of a timestamp column</summary>

`F.min` and `F.max` work on timestamp/date columns just like numeric ones:
```python
df.agg(
    F.to_date(F.min("dateTime")).alias("first_transaction_date"),
    F.to_date(F.max("dateTime")).alias("latest_transaction_date")
)
```
This returns a single-row DataFrame with both boundary dates.

</details>


## Problem 7


<details>
<summary>Hint 1 — most recent date per group</summary>

Group by `customerID` and take the `max` of `dateTime` to get each customer's latest order:
```python
df.groupBy("customerID") \
  .agg(F.to_date(F.max("dateTime")).alias("last_order_date")) \
  .orderBy(F.col("last_order_date").desc())
```

</details>


<details>
<summary>Hint 2 — flipping to find who ordered first</summary>

To find the customer who placed their very **first** order earliest, swap `max` for `min`
and sort ascending:
```python
df.groupBy("customerID") \
  .agg(F.to_date(F.min("dateTime")).alias("first_order_date")) \
  .orderBy(F.col("first_order_date").asc()) \
  .limit(1)   # the very first customer to place an order
```

</details>

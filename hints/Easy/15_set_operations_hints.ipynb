## Problem 1


<details>
<summary>Hint 1 — add a constant source column with F.lit(), then use unionByName</summary>

`F.lit("value")` creates a column where every row has the same constant value:
```python
bakehouse_countries = customers.select("country").distinct()\
    .withColumn("source", F.lit("bakehouse"))

wanderbricks_countries = users.select("country").distinct()\
    .withColumn("source", F.lit("wanderbricks"))

result_1 = bakehouse_countries.unionByName(wanderbricks_countries)
```
`unionByName` matches by column name (safer than `union` which matches by position).

</details>

## Problem 5


<details>
<summary>Hint 1 — select and alias columns to create a matching schema before union</summary>

Both DataFrames must have identical column names before `unionByName`. Use `.select()` with `.alias()` to align them:
```python
from_customers = customers.select(
    F.col("customerID").alias("user_id"),
    F.col("first_name").alias("name"),
    F.col("country"),
    F.lit("bakehouse").alias("source")
)

from_users = users.select(
    F.col("user_id"),
    F.col("full_name").alias("name"),  # check the actual column name
    F.col("country"),
    F.lit("wanderbricks").alias("source")
)

result_5 = from_customers.unionByName(from_users)
```
Run `users.printSchema()` first to confirm the exact column names.

</details>

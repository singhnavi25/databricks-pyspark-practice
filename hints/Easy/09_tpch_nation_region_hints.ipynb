<summary>Hint 1 — collecting multiple values into an array per group</summary>

`F.collect_list()` gathers all values in a group into an array:
```python
joined.groupBy(F.col("r_name").alias("region")) \
      .agg(F.collect_list("n_name").alias("nations"))
```

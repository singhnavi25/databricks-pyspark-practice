## Problem 4


<details>
<summary>Hint 1 — grouping by two extracted date parts</summary>

```python
df.groupBy(
    F.year("o_orderdate").alias("year"),
    F.month("o_orderdate").alias("month")
).agg(
    F.count("*").alias("order_count"),
    F.round(F.avg("o_totalprice"), 2).alias("avg_order_price")
).orderBy("year", "month")
```

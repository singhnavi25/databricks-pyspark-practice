## Problem 1


<details>
<summary>Hint 1 — extract multiple date parts in one select</summary>

Call each date function on the same column inside one `.select()`:
```python
df.select(
    "tpep_pickup_datetime",
    F.year("tpep_pickup_datetime").alias("year"),
    F.month("tpep_pickup_datetime").alias("month"),
    F.dayofmonth("tpep_pickup_datetime").alias("day"),
    F.hour("tpep_pickup_datetime").alias("hour"),
    F.minute("tpep_pickup_datetime").alias("minute")
)
```

</details>

## Problem 2


<details>
<summary>Hint 1 — cast timestamps to long (seconds) before subtracting</summary>

Spark cannot subtract two timestamp columns directly. Cast both to `long` first:
```python
df.withColumn(
    "duration_minutes",
    (F.col("tpep_dropoff_datetime").cast("long") -
     F.col("tpep_pickup_datetime").cast("long")) / 60
)
```
`cast("long")` converts a timestamp to Unix epoch seconds.

</details>

## Problem 1


<details>
<summary>Hint 1 — import types as T and use T.StringType() inside cast()</summary>

The setup cell imports `types as T`. Use it like this:
```python
df.withColumn("card_str", F.col("cardNumber").cast(T.StringType()))
  .withColumn("revenue_double", F.col("totalPrice").cast(T.DoubleType()))
```
You can also use string shortcuts: `.cast("string")` and `.cast("double")`.

## Problem 2


<details>
<summary>Hint 1 — chain withColumnRenamed calls or use select with aliases</summary>

Chain multiple `.withColumnRenamed()` calls:
```python
df.withColumnRenamed("o_orderkey", "order_id")
  .withColumnRenamed("o_custkey", "customer_id")
  .drop("o_comment")
```
Or use a single `.select()` with `.alias()` — that renames and selects in one step:
```python
df.select(
    F.col("o_orderkey").alias("order_id"),
    F.col("o_custkey").alias("customer_id")
)
```

</details>

## Problem 3


<details>
<summary>Hint 1 — chain when().when().otherwise() for multiple categories</summary>

Use `.when()` for each condition and `.otherwise()` as the catch-all:
```python
F.when(F.col("totalPrice") < 20, "Low")
 .when(F.col("totalPrice") <= 100, "Medium")
 .otherwise("High")
```
Spark evaluates conditions top-to-bottom and stops at the first match.
So the second `.when()` only fires when `totalPrice >= 20` (the first condition was false).

</details>

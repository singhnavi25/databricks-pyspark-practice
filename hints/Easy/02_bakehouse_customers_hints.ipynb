## Problem 3


<details>
<summary>Hint 1 — replacing nulls before groupBy</summary>

`F.coalesce` returns the first non-null value:
```python
F.coalesce(F.col("gender"), F.lit("Unknown"))
```
Or use `F.when`:
```python
F.when(F.col("gender").isNull(), "Unknown").otherwise(F.col("gender"))
```

</details>


## Problem 4


<details>
<summary>Hint 1 — combining null check with empty string check</summary>

Use `|` (OR) to combine two filter conditions:
```python
df.filter(
    F.col("state").isNull() | (F.col("state") == "")
)
```
Note: use `&` for AND and `|` for OR — not Python's `and`/`or` keywords.

</details>


## Problem 5


<details>
<summary>Hint 1 — top N with two groupBy columns</summary>

Group by both columns together:
```python
df.groupBy("city", "country") \
  .count() \
  .orderBy(F.col("count").desc()) \
  .limit(3)
```

</details>
